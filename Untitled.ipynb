```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import csv

# Chemin vers ChromeDriver (à modifier selon ton installation)
driver = webdriver.Chrome(executable_path='PROJET_NLP/chromedriver')

# Ouvrir Facebook
driver.get('https://www.facebook.com')

# Pause pour que tu puisses te connecter manuellement
input("Connecte-toi à Facebook, puis appuie sur Entrée ici...")

# Accéder à la page de recherche publique pour UVBF
search_url = 'https://www.facebook.com/search/top?q=UVBF'
driver.get(search_url)

# Attendre quelques secondes pour que la page charge complètement
time.sleep(5)

# Faire défiler la page pour charger plus de publications
scrolls = 5  # Nombre de scrolls pour charger plus de contenu
for i in range(scrolls):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)  # Pause pour que les publications se chargent

# Extraire le contenu de la page avec BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'lxml')

# Fermer le navigateur après avoir obtenu le contenu
driver.quit()

# Trouver les publications publiques sur UVBF (la structure HTML peut changer)
publications = soup.find_all('div', {'class': 'du4w35lb k4urcfbm l9j0dhe7 sjgh65i0'})

# Sauvegarder les données dans un fichier CSV
with open('publications_uvbf.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Date', 'Auteur', 'Texte'])

    # Extraire et écrire chaque publication dans le fichier CSV
    for post in publications:
        try:
            # Extraire le texte du post
            texte_post = post.find('div', {'class': 'ecm0bbzt e5nlhep0 a8c37x1j'}).text
            # Extraire la date et l'auteur si disponible
            date_post = post.find('abbr').text if post.find('abbr') else 'Inconnue'
            auteur_post = post.find('span', {'class': 'fwb'}).text if post.find('span', {'class': 'fwb'}) else 'Inconnu'

            # Écrire les informations dans le fichier CSV
            writer.writerow([date_post, auteur_post, texte_post])
        except AttributeError:
            continue

print("Scraping terminé et les données sont sauvegardées dans 'publications_uvbf.csv'.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import json
from datetime import datetime

def scrape_facebook_posts(keyword, limit=250):
    posts = []
    base_url = f'https://www.facebook.com/search/posts/?q={keyword}'
    page_num = 1

    while len(posts) < limit:
        url = f"{base_url}&page={page_num}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        post_elements = soup.find_all('div', class_='story_body_container')
        for post_element in post_elements:
            post_text = post_element.find('div', class_='story_body').get_text(strip=True)
            post_author = post_element.find('a', class_='author').get_text(strip=True)
            post_date = post_element.find('abbr')['title']
            
            posts.append({
                "source": "Facebook",
                "text": post_text,
                "date": datetime.strptime(post_date, "%Y-%m-%dT%H:%M:%S%z").isoformat(),
                "author": post_author
            })

            if len(posts) >= limit:
                break

        page_num += 1

    return posts

def save_to_file(data, filename):
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        if filename.endswith('.csv'):
            writer = csv.DictWriter(file, fieldnames=["source", "text", "date", "author"])
            writer.writeheader()
            writer.writerows(data)
        elif filename.endswith('.json'):
            json.dump(data, file, ensure_ascii=False, indent=4)

if __name__ == "__main__":
    keyword = "UVBF"
    facebook_data = scrape_facebook_posts(keyword)
    
    save_to_file(facebook_data, "uvbf_facebook_posts.csv")
    save_to_file(facebook_data, "uvbf_facebook_posts.json")
    
    print(f"Scraping terminé. {len(facebook_data)} publications Facebook sauvegardées.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import csv

# Chemin vers ChromeDriver
chrome_driver_path = "chromedriver"  # Mettez à jour selon votre installation

# Options de Chrome pour Selenium
chrome_options = Options()
chrome_options.add_argument("--headless")  # Exécuter Chrome en mode headless, sans interface graphique
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Configurer le driver Chrome avec Selenium
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Accéder à Facebook
driver.get('https://www.facebook.com/search/posts?q=UVBF')

# Pause pour laisser la page se charger
time.sleep(5)

# Scroll down pour charger plus de résultats (facultatif)
for _ in range(3):  # Ajustez le nombre de scrolls selon les besoins
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Trouver les posts (utilisez les bonnes classes CSS après inspection)
posts = driver.find_elements(By.CSS_SELECTOR, "div[data-ad-preview='message']")

# Sauvegarder les données dans un fichier CSV
with open('uvbf_posts.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Texte'])

    for post in posts:
        texte = post.text
        writer.writerow([texte])

# Fermer le driver
driver.quit()
```
